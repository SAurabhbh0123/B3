```python
import csv
def read_csv_to_2d_list(file_path):
    data = []
    with open(file_path, 'r') as file:
        csv_reader = csv.reader(file)
        for row in csv_reader:
            data.append(row)
    return data
file_path = 'Employee.csv'
csv_data = read_csv_to_2d_list(file_path)
for row in csv_data:
    print(row)
```
